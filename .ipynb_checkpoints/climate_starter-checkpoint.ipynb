# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
engine = create_engine("sqlite:///resources/hawaii.sqlite")
inspector = inspect(engine)
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect = True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
temp_mea = Base.classes.measurement
stations = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
col_temp_mea = inspector.get_columns("measurement")
for column in col_temp_mea:
    print(column["name"], column["type"])
    
print("-------------")

col_stations = inspector.get_columns("station")
for column in col_stations:
    print(column["name"], column["type"])
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results
# Calculate the date 1 year ago from the last data point in the database
last_date = session.query(temp_mea.date).order_by(temp_mea.date.desc()).first()
one_year_ago = str(int(last_date[0][:4]) - 1)
last_year_date = one_year_ago + last_date[0][4:]

# Perform a query to retrieve the data and precipitation scores
rain_log = session.query(temp_mea.date, temp_mea.prcp).filter(temp_mea.date >= last_year_date).filter(temp_mea.station == 'USC00519281').order_by(temp_mea.date).all()

# Save the query results as a Pandas DataFrame and set the index to the date column
rain_df = pd.DataFrame(rain_log).rename(columns = {"prcp" : "precipitation"}).set_index("date")
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
rain_df.plot(kind = "line", figsize = (8,5), color="g")

indexes = np.arange(0,351,30)
dates = [rain_df.index[index] for index in indexes]
plt.xticks(np.arange(0,351,30), dates, rotation=90)
plt.ylabel("inches")
plt.title("Precipitation last year record", fontsize=15);
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
rain_df.describe()
```

```python
# Design a query to show how many stations are available in this dataset?
unique_stations = session.query(temp_mea.station).distinct()
[station for station in unique_stations]
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.

#NOTE: Since there are no missing values in the "tobs" column, I decided to evaluate "the most active station" by counting the
#number of "prcp" observations, where there are missing values. In this way, I'm counting complete rows (prcp + tobs).
station_records_count = session.query(temp_mea.station, func.count(temp_mea.prcp)).\
    group_by(temp_mea.station).\
    order_by(func.count(temp_mea.prcp).desc())

[station for station in station_records_count]
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?
all_temp = session.query(func.min(temp_mea.tobs), func.max(temp_mea.tobs), func.avg(temp_mea.tobs)).\
filter(temp_mea.station == station_records_count[0][0])

for temp in all_temp:
    print(f"Min temp is {temp[0]}\n\
Max temp is {temp[1]}\n\
Ave temp is {round(temp[2],2)}")
```

```python
# Choose the station with the highest number of temperature observations.
temp_obs_station = session.query(temp_mea.station, func.count(temp_mea.tobs)).\
    group_by(temp_mea.station).\
    order_by(func.count(temp_mea.tobs).desc())
temp_obs_station[0]
```

```python
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
temp_log = session.query(temp_mea.date, temp_mea.tobs).\
filter(temp_mea.date >= last_year_date).\
filter(temp_mea.station == temp_obs_station[0][0]).\
order_by(temp_mea.date).all()

temp_df = pd.DataFrame(temp_log).set_index("date")

temp_df.plot(kind = "hist", figsize = (8,5), color="orange")
plt.xlabel("Temperature (F)")
plt.figtext(.5,.95,"Temperature last year record", fontsize = 15, ha="center")
plt.figtext(.5,.9, f"Station:{temp_obs_station[0][0]}", ha = "center", fontsize = 10);
```
